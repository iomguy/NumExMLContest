```python
import pandas as pd
import catboost
import numpy as np
import tqdm
from sklearn.model_selection import KFold
from catboost import CatBoostRegressor, FeaturesData, Pool
from matplotlib import pyplot as plt

best_combinations_amount = 10
number_of_tests = 400
common_best_combinations_list = []

for i in tqdm.tqdm_notebook(range(number_of_tests)):
    data_train = pd.read_csv("Results(4000).csv", engine='python', encoding="windows-1251", sep=";", index_col=False,
                             skiprows=2)
    # cols = ["Well_type"]
    # cols_indexes = [data_train.columns.get_loc(c) for c in data_train.columns if c in cols]
    # NNS = data_train.loc[data_train ['column_name'] == some_value]
    nns = data_train.loc[data_train["Well_type"] == 1]
    nns_grp = data_train.loc[data_train["Well_type"] == 2]
    gs = data_train.loc[data_train["Well_type"] == 3]
    gs_grp = data_train.loc[data_train["Well_type"] == 4]
    
    nns_numeric_features_names = ["PSS", "Pprod", "Pinj", "Perm", "h", "a/b", "Poro", "Sw", "VdP", "Pbp", "mo",
                                  "Bo", "po", "mw", "Bw", "Pres"]
    nns_features = nns[nns_numeric_features_names]
    nns_target = nns[["NPV"]]

    nns_grp_numeric_features_names = ["PSS", "Pprod", "Pinj", "Perm", "h", "a/b", "Poro", "Sw", "VdP", "Pbp", "mo",
                                      "Bo", "po", "mw", "Bw", "Pres",
                                      "Lfracprod", "Lfracinj", "Frac_perm", "Wprod", "Winj"]
    nns_grp_features = nns_grp[nns_grp_numeric_features_names]
    nns_grp_target = nns_grp[["NPV"]]

    gs_numeric_features_names = ["PSS", "Pprod", "Pinj", "Perm", "h", "a/b", "Poro", "Sw", "VdP", "Pbp", "mo",
                                  "Bo", "po", "mw", "Bw", "Pres", "Lgsprod", "Lgsinj"]
    gs_features = gs[gs_numeric_features_names]
    gs_target = gs[["NPV"]]

    gs_grp_numeric_features_names = ["PSS", "Pprod", "Pinj", "Perm", "h", "a/b", "Poro", "Sw", "VdP", "Pbp", "mo",
                                      "Bo", "po", "mw", "Bw", "Pres",
                                     "Lfracprod", "Lfracinj", "Frac_perm", "Wprod", "Winj",
                                     "Lgsprod", "Lgsinj", "Nfracprod", "Nfracinj"]
    gs_grp_features = gs_grp[gs_grp_numeric_features_names]
    gs_grp_target = gs_grp[["NPV"]]
    
    # train/test division
    skf = KFold(n_splits=9, shuffle=True)
    # range(1) stops "for" after the 1-st iteration
    for _, (nns_train_index, nns_test_index) \
            in zip(range(1), skf.split(np.array(nns_features), np.array(nns_features))):
        # nns_target.iloc[test_index, :]
        nns_features_train, nns_features_test =\
            nns_features.values[nns_train_index], nns_features.values[nns_test_index]
        nns_target_train, nns_target_test = \
            nns_target.values[nns_train_index], nns_target.values[nns_test_index]

    print("nns train: {}".format(nns_features_train.shape))
    print("nns test: {}".format(nns_features_test.shape))

    for _, (nns_grp_train_index, nns_grp_test_index) \
            in zip(range(1), skf.split(np.array(nns_grp_features), np.array(nns_grp_features))):
        # nns_grp_target.iloc[test_index, :]
        nns_grp_features_train, nns_grp_features_test = \
            nns_grp_features.values[nns_grp_train_index], nns_grp_features.values[nns_grp_test_index]
        nns_grp_target_train, nns_grp_target_test = \
            nns_grp_target.values[nns_grp_train_index], nns_grp_target.values[nns_grp_test_index]

    for _, (gs_train_index, gs_test_index) \
            in zip(range(1), skf.split(np.array(gs_features), np.array(gs_features))):
        # gs_target.iloc[test_index, :]
        gs_features_train, gs_features_test = \
            gs_features.values[gs_train_index], gs_features.values[gs_test_index]
        gs_target_train, gs_target_test = \
            gs_target.values[gs_train_index], gs_target.values[gs_test_index]

    for _, (gs_grp_train_index, gs_grp_test_index) \
            in zip(range(1), skf.split(np.array(gs_grp_features), np.array(gs_grp_features))):
        # gs_grp_target.iloc[test_index, :]
        gs_grp_features_train, gs_grp_features_test = \
            gs_grp_features.values[gs_grp_train_index], gs_grp_features.values[gs_grp_test_index]
        gs_grp_target_train, gs_grp_target_test = \
            gs_grp_target.values[gs_grp_train_index], gs_grp_target.values[gs_grp_test_index]
        
    nns_model = CatBoostRegressor(iterations=400, learning_rate=0.6, depth=6, name="nns", loss_function='RMSE', eval_metric="MAE")
    nns_train_pool = catboost.Pool(nns_features_train.astype(np.float32), nns_target_train[:,0].astype(np.float32))
    nns_validate_pool = catboost.Pool(nns_features_test.astype(np.float32), nns_target_test[:,0].astype(np.float32))
    nns_model.fit(nns_train_pool, eval_set=nns_validate_pool, verbose=False, plot=False, use_best_model=True)
    
    nns_grp_model = CatBoostRegressor(iterations=400, learning_rate=0.6, depth=6, name="nns_grp", loss_function='RMSE', eval_metric="MAE")
    nns_grp_train_pool = catboost.Pool(nns_grp_features_train.astype(np.float32), nns_grp_target_train[:,0].astype(np.float32))
    nns_grp_validate_pool = catboost.Pool(nns_grp_features_test.astype(np.float32), nns_grp_target_test[:,0].astype(np.float32))
    nns_grp_model.fit(nns_grp_train_pool, eval_set=nns_grp_validate_pool, verbose=False, plot=False, use_best_model=True)

    gs_model = CatBoostRegressor(iterations=400, learning_rate=0.6, depth=6, name="gs", loss_function='RMSE', eval_metric="MAE")
    gs_train_pool = catboost.Pool(gs_features_train.astype(np.float32), gs_target_train[:,0].astype(np.float32))
    gs_validate_pool = catboost.Pool(gs_features_test.astype(np.float32), gs_target_test[:,0].astype(np.float32))
    gs_model.fit(gs_train_pool, eval_set=gs_validate_pool, verbose=False, plot=False, use_best_model=True)

    gs_grp_model = CatBoostRegressor(iterations=400, learning_rate=0.6, depth=6, name="gs_grp", loss_function='RMSE', eval_metric="MAE")
    gs_grp_train_pool = catboost.Pool(gs_grp_features_train.astype(np.float32), gs_grp_target_train[:,0].astype(np.float32))
    gs_grp_validate_pool = catboost.Pool(gs_grp_features_test.astype(np.float32), gs_grp_target_test[:,0].astype(np.float32))
    gs_grp_model.fit(gs_grp_train_pool, eval_set=gs_grp_validate_pool, verbose=False, plot=False, use_best_model=True)

    common_best_combinations_amounts = []
    
    print("nns best npv combination")
    arr = pd.DataFrame(np.column_stack((nns_model.predict(nns_validate_pool), nns_target_test[:,0])))
    print("NumEX:")
    best_npv_combination_NumEX = np.sort(((arr.sort_values([0], ascending=False))[0:best_combinations_amount]).index)
    print(best_npv_combination_NumEX)
    print("ML:")
    best_npv_combination_ML = np.sort(((arr.sort_values([1], ascending=False))[0:best_combinations_amount]).index)
    print(best_npv_combination_ML)
    print("Common combinations:")
    common_best_combinations = set(best_npv_combination_NumEX).intersection(set(best_npv_combination_ML))
    print("{0} from {1}".format(len(common_best_combinations), best_combinations_amount))    
    common_best_combinations_amounts.append(len(common_best_combinations))
    
    print("nns_grp best npv combination")
    arr = pd.DataFrame(np.column_stack((nns_grp_model.predict(nns_grp_validate_pool), nns_grp_target_test[:,0])))
    print("NumEX:")
    best_npv_combination_NumEX = np.sort(((arr.sort_values([0], ascending=False))[0:best_combinations_amount]).index)
    print(best_npv_combination_NumEX)
    print("ML:")
    best_npv_combination_ML = np.sort(((arr.sort_values([1], ascending=False))[0:best_combinations_amount]).index)
    print(best_npv_combination_ML)
    print("Common combinations:")
    common_best_combinations = set(best_npv_combination_NumEX).intersection(set(best_npv_combination_ML))
    print("{0} from {1}".format(len(common_best_combinations), best_combinations_amount))
    common_best_combinations_amounts.append(len(common_best_combinations))
    
    print("gs best npv combination")

    arr = pd.DataFrame(np.column_stack((gs_model.predict(gs_validate_pool), gs_target_test[:,0])))
    print("NumEX:")
    best_npv_combination_NumEX = np.sort(((arr.sort_values([0], ascending=False))[0:best_combinations_amount]).index)
    print(best_npv_combination_NumEX)
    print("ML:")
    best_npv_combination_ML = np.sort(((arr.sort_values([1], ascending=False))[0:best_combinations_amount]).index)
    print(best_npv_combination_ML)
    print("Common combinations:")
    common_best_combinations = set(best_npv_combination_NumEX).intersection(set(best_npv_combination_ML))
    print("{0} from {1}".format(len(common_best_combinations), best_combinations_amount))
    common_best_combinations_amounts.append(len(common_best_combinations))
    
    print("gs_grp best npv combination")
    arr = pd.DataFrame(np.column_stack((gs_grp_model.predict(gs_grp_validate_pool), gs_grp_target_test[:,0])))
    print("NumEX:")
    best_npv_combination_NumEX = np.sort(((arr.sort_values([0], ascending=False))[0:best_combinations_amount]).index)
    print(best_npv_combination_NumEX)
    print("ML:")
    best_npv_combination_ML = np.sort(((arr.sort_values([1], ascending=False))[0:best_combinations_amount]).index)
    print(best_npv_combination_ML)
    print("Common combinations:")
    common_best_combinations = set(best_npv_combination_NumEX).intersection(set(best_npv_combination_ML))
    print("{0} from {1}".format(len(common_best_combinations), best_combinations_amount))
    common_best_combinations_amounts.append(len(common_best_combinations))
    common_best_combinations_list.append(common_best_combinations_amounts)    
```

```python
common_best_combinations_arr = np.array(common_best_combinations_list)
```

```python
from scipy.stats import norm

# plot the histogram
plt.figure()
plt.grid(zorder=0)
plt.hist(common_best_combinations_arr[:,0], bins=range(0, 12, 1),
         weights=np.ones(len(common_best_combinations_arr[:,0])),
         edgecolor='black', linewidth=0.9, density=True,
         alpha = 0.6,
         color='g',
         zorder=2)
plt.xticks(range(0, best_combinations_amount+1, 1))
# plt.yticks(range(0, number_of_tests+1, 5))
plt.xlabel("Number of proper {} best combinations".format(best_combinations_amount))
plt.ylabel("Number of appearances from {} tests".format(number_of_tests))
plt.title("nns best results")

# fit a normal distribution to the data
nns_mu, nns_std = norm.fit(common_best_combinations_arr[:,0])

# plot the normal distribution histogram
x = np.linspace(nns_mu - 3*nns_std, nns_mu + 3*nns_std, 200)
p = norm.pdf(x, nns_mu, nns_std)
plt.plot(x, p, 'k', linewidth=2, zorder=3)

plt.savefig('nns_bests_histogram.png', format='png', dpi=300)
plt.show()

print("nns mean: {}".format(np.mean(common_best_combinations_arr[:,0])))
print("nns std: {}".format(np.std(common_best_combinations_arr[:,0])))
```
